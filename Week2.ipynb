## 5 Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

width, height = 400, 400

canvas = np.ones((height, width, 3), dtype=np.uint8) * 255

def random_color():
    return np.random.randint(0, 256, size=3).tolist()

num_boxes = 5

for _ in range(num_boxes):
    box_width = np.random.randint(20, 100)
    box_height = np.random.randint(20, 100)

    x = np.random.randint(0, width - box_width)
    y = np.random.randint(0, height - box_height)

    color = random_color()

    cv.rectangle(canvas, (x, y), (x + box_width, y + box_height), color, -1)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
plt.title('Gambar dengan 5 Kotak Acak')
plt.axis('off')
plt.show()
```

## 7 Buat rectangle dan circle pada bagian wajah dari image foto anda saat beraktifitas (bukan pasfoto).

```python

image_path = '/content/drive/MyDrive/PCVK_MM/Images/Waduh.jpg'
img = cv.imread(image_path)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

for (x, y, w, h) in faces:
    cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    center_x = x + w // 2
    center_y = y + h // 2
    radius = min(w, h) // 4
    cv.circle(img, (center_x, center_y), radius, (255, 0, 0), 2)

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Wajah dengan Rectangle dan Circle')
plt.axis('off')
plt.show()
```

## 8 Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!  

```python
def add_rectangle_to_blue_channel(imgB):
    blue_channel = imgB[:, :, 0]

    height, width = blue_channel.shape
    rect_size = 50
    top_left_x = 0
    top_left_y = height - rect_size

    blue_channel[top_left_y:, top_left_x:top_left_x + rect_size] = 255

    imgB[:, :, 0] = blue_channel
    return imgB

image_path2 = '/content/drive/MyDrive/PCVK_MM/Images/kitten01.jpg'
img = cv.imread(image_path2)

modified_img = add_rectangle_to_blue_channel(img)

imgB = cv.cvtColor(modified_img, cv.COLOR_BGR2RGB)

plt.imshow(imgB)
plt.title('Rectangle pada Channel Blue')
plt.axis('off')
plt.show()
```
