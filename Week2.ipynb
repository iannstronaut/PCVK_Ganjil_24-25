## Langkah 4

Menampilkan bentuk Geometri 2D dari OpenCV

Diawali dengan pembuatan black image dengan tipe data int16

```python
black_img = np.zeros((512, 512, 3), dtype=np.int16)
plt.imshow(black_img)
```

menampilkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
cv.rectangle(black_img, pt1=(384, 0), pt2=(510, 150), color =(0, 255, 0), thickness=10)
plt.imshow(black_img)
```

menambahkan bentuk persegi dengan cara yang sama

```python
cv.rectangle(black_img, pt1=(200, 200), pt2=(300, 300), color =(0, 0, 255), thickness=15)
plt.imshow(black_img)
```

menambahkan bentuk lingkaran sesuai radius yang dibuat

```python
cv.circle(black_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)
```

menambahkan garis sesuai koordinat

```python
cv.circle(black_img, center=(400, 400), radius=50, color=(255, 0, 0), thickness=-1)
cv.line(black_img, pt1=(0, 0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)
```

menambahkan text dengan font yang telah tertulis dan ukuran yang ditentukan

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

pembuatan Black Image kembali dengan tipe data int32

```python
black_img2 = np.zeros((512, 512, 3), dtype=np.int32)
plt.imshow(black_img2)
```

kode inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vertices
```

reshape array

```python
pts = vertices.reshape((-1, 1, 2))
pts
```

menambahkan polyline pada black image int32

```python
cv.polylines(black_img2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_img2)
```

### Soal 1

Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?

Perbedaan antara gambar yang ditampilkan tanpa menggunakan Matplotlib dan dengan Matplotlib terletak pada cara visualisasi gambar, terutama dalam hal penanganan warna dan pengaturan sumbu. Matplotlib menggunakan format warna RGB (Red, Green, Blue) secara default. Jika gambar Anda masih dalam format BGR (seperti pada OpenCV), warnanya akan tampil salah (biru terlihat sebagai merah, dan sebaliknya).

### Soal 2

Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?

Pembuatan black image (gambar hitam) dalam berbagai tipe data, seperti int16 dan int32, mempengaruhi beberapa aspek terkait representasi gambar dan penggunaan memori. Untuk gambar hitam, perbedaan dalam tipe data biasanya tidak terlihat secara visual karena semua piksel diset ke 0. Namun, saat Anda memproses gambar lebih lanjut, perbedaan dalam tipe data dapat mempengaruhi hasilnya.
